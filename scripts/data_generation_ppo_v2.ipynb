```python
import pandas as pd
import numpy as np
train=pd.read_csv('../data/train_raw.csv')
```

```python
reshaped_df = pd.DataFrame({
    '질문': pd.concat([train['질문_1']] * 5 + [train['질문_2']] * 5).reset_index(drop=True),
    '답변': pd.concat([train['답변_1'], train['답변_2'] ,train['답변_3'], train['답변_4'],train['답변_5']]*2).reset_index(drop=True)
})
reshaped_df.shape
```

```python
train_raw_dataset = []

for i in range(len(train)):
    row = train.iloc[i]
    q1 = row['질문_1']
    q2 = row['질문_2']

    a1 = row['답변_1']
    a2 = row['답변_2']
    a3 = row['답변_3']
    a4 = row['답변_4']
    a5 = row['답변_5']

    train_raw_dataset.append(
        {
            '질문':q1,
            '답변':[a1,a2,a3,a4,a5]
        }
    )
    train_raw_dataset.append(
        {
            '질문':q2,
            '답변':[a1,a2,a3,a4,a5]
        }
    )
    
```

```python
import itertools
import random

# 같은 카테고리 안에서 질문 2개&그에 따른 답변 2개씩 조합
setlist = []

random.seed(0)
qs = ['질문_1', '질문_2']
answer_cols = ['답변_1','답변_2','답변_3','답변_4','답변_5']
categories = train['category'].unique()

# 1. 카테고리별로 질문 2개씩 조합
for category in categories:
    category_idx = train[train['category'] == category]['id']

    # 2. row 2개씩 조합
    question_combinations = list(itertools.combinations(category_idx, 2))
    lens = len(question_combinations)

    for idx, (i,j) in enumerate(question_combinations):
        # 3. 질문 2개에 대한 답변 2개씩 조합
        for qi1, qi2 in itertools.product(qs, qs):
            if random.random() > 0.01:
                continue
            q1 = train[train['id'] == i][qi1].values[0]
            q2 = train[train['id'] == j][qi2].values[0]

            answers = []
            for a1_col in answer_cols:
                for a2_col in answer_cols:
                    answers.append(f"{train[train['id'] == i][a1_col].values[0]} {train[train['id'] == j][a2_col].values[0]}")
            
            setlist.append({
                    '질문': f'{q1} {q2}',
                    '답변': answers
            })

len(setlist) 
```

```python
raw_dataset = {'data':train_raw_dataset}
concat_dataset = {'data':setlist}
print(len(raw_dataset['data']), len(concat_dataset['data']))
```
